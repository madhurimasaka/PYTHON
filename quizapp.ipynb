```python
import tkinter as tk
from tkinter import messagebox

class QuizApp:
    def __init__(self):
        self.quiz_data = [
            {
                "question": "What is the backend language?",
                "options": ["HTML", "CSS", "JS", "PYTHON"],
                "correct_answer": 3
            },
            {
                "question": "What is the correct syntax to write a variable?",
                "options": ["2var", "var", "__1var", "$var"],
                "correct_answer": 1
            },
            {
                "question": "Who invented Python?",
                "options": ["Ford", "Einstein", "Guido van Rossum", "Galileo"],
                "correct_answer": 2
            }
        ]
        self.current_question_index = 0
        self.score = 0

        self.window = tk.Tk()
        self.window.title("Quiz App")

        self.question_label = tk.Label(self.window, text="", font=("Arial", 16))
        self.question_label.pack(pady=20)

        self.options_frame = tk.Frame(self.window)
        self.options_frame.pack(pady=20)

        self.option_buttons = []
        for i in range(4):
            button = tk.Button(self.options_frame, text="", width=50, font=("Arial", 14), command=lambda i=i: self.check_answer(i))
            button.pack(pady=5)
            self.option_buttons.append(button)

        self.next_button = tk.Button(self.window, text="Next Question", command=self.next_question, state=tk.DISABLED)
        self.next_button.pack(pady=20)

        self.load_question()

    def load_question(self):
        question_data = self.quiz_data[self.current_question_index]
        self.question_label.config(text=question_data["question"])
        for i, option in enumerate(question_data["options"]):
            self.option_buttons[i].config(text=option, state=tk.NORMAL)
        self.next_button.config(state=tk.DISABLED)

    def check_answer(self, selected_option_index):
        correct_answer_index = self.quiz_data[self.current_question_index]["correct_answer"]
        if selected_option_index == correct_answer_index:
            self.score += 1
            messagebox.showinfo("Correct", "That's the correct answer!")
        else:
            messagebox.showerror("Incorrect", "Sorry, that's the wrong answer.")
        
        # Disable all option buttons
        for button in self.option_buttons:
            button.config(state=tk.DISABLED)
        
        self.next_button.config(state=tk.NORMAL)

    def next_question(self):
        self.current_question_index += 1
        if self.current_question_index < len(self.quiz_data):
            self.load_question()
        else:
            messagebox.showinfo("Quiz Finished", f"You've completed the quiz! Your score is {self.score}/{len(self.quiz_data)}.")
            self.window.destroy()

    def start_quiz(self):
        self.window.mainloop()

quiz_app = QuizApp()
quiz_app.start_quiz()
```
